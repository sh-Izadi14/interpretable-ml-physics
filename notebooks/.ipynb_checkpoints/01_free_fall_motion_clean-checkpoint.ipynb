# Interpreting Free-Fall Motion with Linear Regression in PyTorch

This notebook explores whether machine learning can recover physically meaningful parameters from a simple free-fall motion equation. It also tests how normalization affects learning when the underlying relationship is quadratic.


## Why This Problem?

After learning basic linear regression in PyTorch, I realized that recovering arbitrary weights and biases lacked physical meaning. This project asks:

- Can ML recover interpretable parameters from real-world physics?
- How does normalization affect learning in a quadratic system?
- Can linear regression approximate a quadratic relationship when features are engineered?

The equation used:
$$x(t) = -\frac{1}{2}gt^2 + v_0 t + x_0$$


## Approching the Problem using MAE Loss Function Without Normalization

### Creating Data

```python
import torch
from torch import nn
import matplotlib.pyplot as plt
```

```python
n_samples = 100
start = 0
end = 10
steps = (end - start)/100
t = torch.arange(start, end, steps).unsqueeze(dim=1)
# Feature engineering: using t and t² to allow LR to capture quadratic behavior
# X = torch.cat([t, t**2, torch.ones_like(t)], dim=1)  # shape: [100, 3]
X = torch.cat([t, t**2], dim=1)  # shape: [100, 2]  the nn.Linear() handles the bias term.
## Simulating free-fall motion with gravity g = 9.8 m/s² 
a = 4.9 # constant
b = 20 # V0 m/s
c = 300 # x0 m (bias term)
y = -a*t**2 + b*t + c
```

### Spliting the Data

```python
train_size = int(0.8 * len(t))
X_train = X[:train_size]
y_train = y[:train_size]
X_test = X[train_size:]
y_test = y[train_size:]

len(X_train), len(y_train), len(X_test), len(y_test)
```

### Defining the Linear Model Architecture	

```python
class LinearRegressionModel(nn.Module):
    def __init__(self):
        super().__init__()
        # Use nn.Linear() for creating the model parameters
        self.linear_layer = nn.Linear(in_features=2, out_features=1)
    
    # Define the forward computation (input data x flows through nn.Linear())
    def forward(self, x: torch.Tensor) -> torch.Tensor:
        return self.linear_layer(x)
```

```python
# Set manual seed since nn.Parameter are randomly initialized
torch.manual_seed(42)

# Create an instance of the model (this is a subclass of nn.Module that contains nn.Parameter(s))
model = LinearRegressionModel()

# List named parameters 
print('\n', model.state_dict())
```

### Visualizing the Initial Perdiction (Random Parameters)

```python
import sys
sys.path.append('../utils')
from plot_predictions import plot_predictions
# Make predictions with model
with torch.inference_mode(): 
    y_preds = model(X_test)

plot_predictions(train_data=X_train, 
                     train_labels=y_train, 
                     test_data=X_test, 
                     test_labels=y_test, 
                     predictions=y_preds)
```

### Selection of the Optimizer and the Loss Function

```python
# Create the loss function
loss_fn = nn.L1Loss() # MAE loss is same as L1Loss

# Create the optimizer
optimizer = torch.optim.SGD(params=model.parameters(), lr=0.01)
```

### Train, Test Loop

```python
torch.manual_seed(42)

# Create models directory if it doesn't exist
import os
os.makedirs("../models", exist_ok=True)

epochs = 10000

# Set the number of epochs (how many times the model will pass over the training data)
from collections import deque

window_size = 100  # Number of recent epochs to track
patience = 1000    # How long to wait after plateau
loss_window = deque(maxlen=window_size)
best_loss = float('inf')
trigger_times = 0

# Create empty loss lists to track values
train_loss_values = []
test_loss_values = []
best_loss_values = []
epoch_count = []

for epoch in range(epochs):
    ### Training
    
    # Put model in training mode (this is the default state of a model)
    model.train()
    
    # 1. Forward pass on train data using the forward() method inside 
    y_pred = model(X_train)
    # print(y_pred)
    
    # 2. Calculate the loss (how different are our models predictions to the ground truth)
    loss = loss_fn(y_pred, y_train)
    
    # 3. Zero grad of the optimizer
    optimizer.zero_grad()
    
    # 4. Loss backwards
    loss.backward()
    
    # 5. Progress the optimizer
    optimizer.step()
    
    ### Testing
    
    # Put the model in evaluation mode
    model.eval()
    
    with torch.inference_mode():
      # 1. Forward pass on test data
      test_pred = model(X_test)
    
      # 2. Caculate loss on test data
      test_loss = loss_fn(test_pred, y_test.type(torch.float)) # predictions come in torch.float datatype, so comparisons need to be done with tensors of the same type
    
    # Early stopping logic     
    current_loss = test_loss.item()
    loss_window.append(current_loss)

    if current_loss < best_loss:
        best_loss = current_loss
        best_loss_values.append(best_loss)
        torch.save(model.state_dict(), '../models/best_L1_model.pth')
        trigger_times = 0
    else:
        # Check if current loss is consistently worse than the average of recent losses
        if len(loss_window) == window_size and current_loss > sum(loss_window)/window_size:
            trigger_times += 1
            if trigger_times >= patience:
                print(f"Early stopping at epoch {epoch}")
                break


    # Print out what's happening
    if epoch % 100 == 0:
        epoch_count.append(epoch)
        train_loss_values.append(loss.detach().numpy())
        test_loss_values.append(test_loss.detach().numpy())
        print(f"Epoch: {epoch} | MAE Train Loss: {loss} | MAE Test Loss: {test_loss} ")
        print(f"Best Loss Value: {best_loss}")
```

### Plot the Loss Curves

```python
plt.plot(epoch_count, train_loss_values, label="Train loss")
plt.plot(epoch_count, test_loss_values, label="Test loss")
plt.plot(best_loss_values, '*', label="Best loss")
plt.title("Training and test loss curves")
plt.ylabel("Loss")
plt.xlabel("Epochs")
plt.legend();
```

## Approching the Problem using MAE Loss Function With Normalization

```python
import torch
from torch import nn
import numpy as np
```

### Creating, Normalizing and Splitting the Data

```python
# Creating
start = 0 
end = 10
step = 0.1
a = 4.9
b = 30
c = 300
t = torch.arange(start, end, step).unsqueeze(dim=1)
X = torch.cat([t, t**2], dim=1)
y = -a*t**2 + b*t + c

# Normalization
X_mean = X.mean(dim=0)
X_std = X.std(dim=0)
X_normalized = (X -X_mean) / (X_std + 1e-8)


y_mean = X.mean()
y_std = X.std()
y_normalized = (y -y_mean) / (y_std + 1e-8)

# Splitting
train_size = int(0.8 * len(t))
X_train = X_normalized[:train_size]
y_train = y_normalized[:train_size]
X_test = X_normalized[train_size:]
y_test = y_normalized[train_size:]
```

### Defining the Linear Model Architecture	

```python
class LinearRegressionModel(nn.Module):
    def __init__(self):
        super().__init__()
        # Use nn.Linear() for creating the model parameters
        self.linear_layer = nn.Linear(in_features=2, out_features=1)
    
    # Define the forward computation (input data x flows through nn.Linear())
    def forward(self, x: torch.Tensor) -> torch.Tensor:
        return self.linear_layer(x)
```

```python
# Set manual seed since nn.Parameter are randomly initialized
torch.manual_seed(42)

# Create an instance of the model (this is a subclass of nn.Module that contains nn.Parameter(s))
model = LinearRegressionModel()

# List named parameters 
print(model.state_dict())
```

### Selecting the Optimizer and Loss Function

```python
# Create the loss function
loss_fn = nn.L1Loss() # MAE loss is same as L1Loss

# Create the optimizer
optimizer = torch.optim.SGD(params=model.parameters(), lr=0.01)
```

### Train, Test Loop

```python
torch.manual_seed(42)


epochs = 10000

# Set the number of epochs (how many times the model will pass over the training data)
from collections import deque

window_size = 100  # Number of recent epochs to track
patience = 1000    # How long to wait after plateau
loss_window = deque(maxlen=window_size)
best_loss = float('inf')
trigger_times = 0

# Create empty loss lists to track values
train_loss_values = []
test_loss_values = []
best_loss_values = []
epoch_count = []

for epoch in range(epochs):
    ### Training
    
    # Put model in training mode (this is the default state of a model)
    model.train()
    
    # 1. Forward pass on train data using the forward() method inside 
    y_pred = model(X_train)
    # print(y_pred)
    
    # 2. Calculate the loss (how different are our models predictions to the ground truth)
    loss = loss_fn(y_pred, y_train)
    
    # 3. Zero grad of the optimizer
    optimizer.zero_grad()
    
    # 4. Loss backwards
    loss.backward()
    
    # 5. Progress the optimizer
    optimizer.step()
    
    ### Testing
    
    # Put the model in evaluation mode
    model.eval()
    
    with torch.inference_mode():
      # 1. Forward pass on test data
      test_pred = model(X_test)
    
      # 2. Caculate loss on test data
      test_loss = loss_fn(test_pred, y_test.type(torch.float)) # predictions come in torch.float datatype, so comparisons need to be done with tensors of the same type
    
    # Early stopping logic     
    current_loss = test_loss.item()
    loss_window.append(current_loss)

    if current_loss < best_loss:
        best_loss = current_loss
        best_loss_values.append(best_loss)
        torch.save(model.state_dict(), '../models/best_L1_model_normalized.pth')
        trigger_times = 0
    else:
        # Check if current loss is consistently worse than the average of recent losses
        if len(loss_window) == window_size and current_loss > sum(loss_window)/window_size:
            trigger_times += 1
            if trigger_times >= patience:
                print(f"Early stopping at epoch {epoch}")
                break


    # Print out what's happening
    if epoch % 100 == 0:
        epoch_count.append(epoch)
        train_loss_values.append(loss.detach().numpy())
        test_loss_values.append(test_loss.detach().numpy())
        print(f"Epoch: {epoch} | MAE Train Loss: {loss} | MAE Test Loss: {test_loss} ")
        print(f"Best Loss Value: {best_loss}")
```

### Plot the Loss Curves

```python
plt.plot(epoch_count, train_loss_values, label="Train loss")
plt.plot(epoch_count, test_loss_values, label="Test loss")
plt.plot(best_loss_values, '*', label="Best loss")
plt.title("Training and test loss curves")
plt.ylabel("Loss")
plt.xlabel("Epochs")
plt.legend();
```

## Approaching the Problem using MSE Loss Function with Normalization

### 🧠 Why We Skip MSE Without Normalization

In earlier experiments, it became clear that **normalization dramatically improves training stability and convergence**, especially when using loss functions that amplify error magnitudes.

Given that:
- **MSE (Mean Squared Error)** squares the error, making it highly sensitive to large values
- Unnormalized data can lead to numerical instability (e.g., NaNs or exploding gradients)
- Normalization consistently led to faster convergence and more interpretable parameter recovery

We conclude that testing MSE **without normalization** is unnecessary and potentially misleading. Instead, we proceed directly to the normalized setup, which reflects best practices in both machine learning and scientific modeling.

> This decision streamlines the experiment and focuses our attention on meaningful results, not avoidable instability.


```python
import torch
from torch import nn
import numpy as np
```

### Creating, Normalizing and Splitting the Data

```python
# Creating
start = 0 
end = 10
step = 0.1
a = 4.9
b = 30
c = 300
t = torch.arange(start, end, step).unsqueeze(dim=1)
X = torch.cat([t, t**2], dim=1)
y = -a*t**2 + b*t + c

# Normalization
X_mean = X.mean(dim=0)
X_std = X.std(dim=0)
X_normalized = (X -X_mean) / (X_std + 1e-8)


y_mean = X.mean()
y_std = X.std()
y_normalized = (y -y_mean) / (y_std + 1e-8)

# Splitting
train_size = int(0.8 * len(t))
X_train = X_normalized[:train_size]
y_train = y_normalized[:train_size]
X_test = X_normalized[train_size:]
y_test = y_normalized[train_size:]
```

### Defining the Linear Model Architecture	

```python
class LinearRegressionModel(nn.Module):
    def __init__(self):
        super().__init__()
        # Use nn.Linear() for creating the model parameters
        self.linear_layer = nn.Linear(in_features=2, out_features=1)
    
    # Define the forward computation (input data x flows through nn.Linear())
    def forward(self, x: torch.Tensor) -> torch.Tensor:
        return self.linear_layer(x)
```

### Selecting the Optimizer and Loss Function

```python
# Create the loss function
loss_fn = nn.MSELoss() # MSE loss Function

# Create the optimizer
optimizer = torch.optim.SGD(params=model.parameters(), lr=0.01)
```

### Train, Test Loop

```python
torch.manual_seed(42)


epochs = 10000

# Set the number of epochs (how many times the model will pass over the training data)
from collections import deque

window_size = 100  # Number of recent epochs to track
patience = 1000    # How long to wait after plateau
loss_window = deque(maxlen=window_size)
best_loss = float('inf')
trigger_times = 0

# Create empty loss lists to track values
train_loss_values_mse = []
test_loss_values_mse = []
best_loss_values_mse = []
epoch_count_mse = []

for epoch in range(epochs):
    ### Training
    
    # Put model in training mode (this is the default state of a model)
    model.train()
    
    # 1. Forward pass on train data using the forward() method inside 
    y_pred = model(X_train)
    # print(y_pred)
    
    # 2. Calculate the loss (how different are our models predictions to the ground truth)
    loss = loss_fn(y_pred, y_train)
    
    # 3. Zero grad of the optimizer
    optimizer.zero_grad()
    
    # 4. Loss backwards
    loss.backward()
    
    # 5. Progress the optimizer
    optimizer.step()
    
    ### Testing
    
    # Put the model in evaluation mode
    model.eval()
    
    with torch.inference_mode():
      # 1. Forward pass on test data
      test_pred = model(X_test)
    
      # 2. Caculate loss on test data
      test_loss = loss_fn(test_pred, y_test.type(torch.float)) # predictions come in torch.float datatype, so comparisons need to be done with tensors of the same type
    
    # Early stopping logic     
    current_loss = test_loss.item()
    loss_window.append(current_loss)

    if current_loss < best_loss:
        best_loss = current_loss
        best_loss_values_mse.append(best_loss)
        torch.save(model.state_dict(), '../models/best_mse_model_normalized.pth')
        trigger_times = 0
    else:
        # Check if current loss is consistently worse than the average of recent losses
        if len(loss_window) == window_size and current_loss > sum(loss_window)/window_size:
            trigger_times += 1
            if trigger_times >= patience:
                print(f"Early stopping at epoch {epoch}")
                break


    # Print out what's happening
    if epoch % 100 == 0:
        epoch_count_mse.append(epoch)
        train_loss_values_mse.append(loss.detach().numpy())
        test_loss_values_mse.append(test_loss.detach().numpy())
        print(f"Epoch: {epoch} | MSE Train Loss: {loss} | MSE Test Loss: {test_loss} ")
        print(f"Best Loss Value: {best_loss}")
```

### Plot the Loss Curves

```python
plt.plot(epoch_count_mse, train_loss_values_mse, label="Train loss")
plt.plot(epoch_count_mse, test_loss_values_mse, label="Test loss")
plt.plot(best_loss_values_mse, '*', label="Best loss")
plt.title("Training and test loss curves")
plt.ylabel("Loss")
plt.xlabel("Epochs")
plt.legend();
```

```python
import sys
sys.path.append('../utils')
from denormalizing import denormalize_parameters
state_dict = torch.load('../models/best_MSE_model_normalized.pth', weights_only=True)

weights = state_dict['linear_layer.weight']
bias = state_dict['linear_layer.bias']
X_mean = X_mean
X_std = X_std
y_mean = y_mean
y_std = y_std

a, b, c = denormalize_parameters(weights, bias, X_mean, X_std, y_mean, y_std)
print(f"Recovered physical parameters:\na = {a:.2f}, b = {b:.2f}, c = {c:.2f}")
```

## 🔍 Loss Function & Normalization Insights

During experimentation, I observed the following:

- **MAE (Mean Absolute Error)** without normalization produced a reasonably accurate model, but required **~8700 epochs** and resulted in **moderate loss values**.
- **MAE with normalization** showed improved convergence behavior, requiring fewer epochs and yielding more stable loss values — though still slower than MSE, it preserved robustness and interpretability.
- **MSE (Mean Squared Error)** without normalization led to **NaN values**, likely due to squaring large errors exceeding `float32` limits.
- After applying **normalization**, MSE performed significantly better, converging in **~4000 epochs** with **very low training and test losses**, and recovering physically meaningful parameters.

### 🧠 What This Highlights

- **Normalization** is essential for numerical stability, especially when using loss functions sensitive to scale.
- **MAE** offers robustness but slower convergence, while **MSE** benefits more dramatically from normalization.
- Choosing the right combination of loss function and preprocessing is key to balancing **accuracy**, **efficiency**, and **interpretability**.

> These insights guided the decision to skip MSE without normalization in future experiments, focusing instead on setups that reflect best practices and yield meaningful results.
